Task:
● Product Catalogue Cleaner: A supplier sends a raw product list with mixed types,
duplicates, and missing values. Clean the data, convert all prices to float, group products
into a dictionary by category, and print a neat summary.

```python
# 1. Raw Data (from supplier)

raw_products = [
    {"name": "Laptop", "category": "Electronics", "price": "1200.50"},
    {"name": "Mouse", "category": "Electronics", "price": 25},
    {"name": "Laptop", "category": "Electronics", "price": "1200.50"},
    {"name": "Desk", "category": "Furniture", "price": "150.00"},
    {"name": "Chair", "category": "Furniture", "price": None},
    {"name": "Headphones", "category": "Electronics", "price": "N/A"},
    {"name": "Pen", "category": "Stationery", "price": "5.75"},
    {"name": "Notebook", "category": "Stationery", "price": "2.50"},
    {"name": "Laptop", "price": "75000", "category": "Electronics"},
    {"name": "Mobile", "price": 50000, "category": "Electronics"},
    {"name": "Chair", "price": "3000", "category": "Furniture"},
    {"name": "Laptop", "price": "75000", "category": "Electronics"},
    {"name": "Table", "price": None, "category": "Furniture"},
    {"name": "Pen", "price": "50", "category": None},
    {"name": "Notebook", "price": "120", "category": "Stationery"}
]

# 2. Data Structures Initialize Karna
cleaned_catalog = {}  # dictionary for Final grouped data
seen_products = set() # for tracking Duplicates

# 3. Data Cleaning and Grouping  Process
for item in raw_products:
    name = item.get("name")
    category = item.get("category")
    raw_price = item.get("price")

    # if there is any product name in 'seen_products' set , then skip (Duplicate check)
    if name in seen_products:
        continue

    # try to convert prince in floating
    try:
        # Type casting: string ya int ko float banayega
        clean_price = float(raw_price)
    except (ValueError, TypeError):
        # Agar price 'None' ya 'N/A' hui, toh error aayega. Hum aisi product ko skip kar denge.
        continue

    # Agar code yahan tak pohanch gaya, matlab data valid aur naya hai.
    # Product ko 'seen' set mein daal dein taake dobara na aaye
    seen_products.add(name)

    # Dictionary mein category ke hisaab se group karein (setdefault ka use)
    # Agar category pehle se nahi hai toh khali list [] bana dega, warna purani list mein append karega
    cleaned_catalog.setdefault(category, []).append({"name": name, "price": clean_price})


# 4. Cleaned Summary to print
print("---  CLEANED PRODUCT CATALOGUE ---\n")

for category, items in cleaned_catalog.items():
    print(f"Category: {category.upper()}")
    total_price = 0

    for item in items:
        print(f"  - {item['name']} : Rs. {item['price']}")
        total_price += item["price"]

    print(f"  Total Products: {len(items)}")
    print(f"  Total Value: Rs. {total_price}\n")
```

```python
import os
import csv

# CONFIGURATION

folder = "student_records"
PASS_MARKS = 50  # Students need 50 or above to pass

# These are ALL the files we expect (including one that doesn't exist)
expected_files = [
    "correct_data.csv",
    "correct_data1.csv",
    "section_c.csv",       # This file does NOT exist (missing)
    "corrupted_file.csv",   # This file is broken
    "invalid_marks.csv",    # This file has bad data
]


# LISTS TO STORE RESULTS

passed_students = []    # Students who scored >= 50
failed_students = []    # Students who scored < 50
errors_log = []         # All errors/problems we encounter


# PROCESS EACH FILE

for filename in expected_files:
    filepath = f"{folder}/{filename}"
    print(f"\n Processing: {filename}")
    print("-" * 40)

    try:
        # Try to open the file
        with open(filepath, "r") as f:
            reader = csv.reader(f)

            # Read the header row
            header = next(reader)

            # Check if header looks correct
            if "Name" not in header or "Marks" not in header:
                raise ValueError(f"File has wrong headers: {header}")

            # Find which column has Name and Marks
            name_index = header.index("Name")
            marks_index = header.index("Marks")

            # Read each student row
            for line_number, row in enumerate(reader, start=2):
                try:
                    # Get student name
                    name = row[name_index].strip()

                    # Try converting marks to a number
                    marks = int(row[marks_index])

                    # Check pass or fail
                    if marks >= PASS_MARKS:
                        passed_students.append({
                            "Name": name,
                            "Marks": marks,
                            "Status": "PASS",
                            "File": filename
                        })
                        print(f"   {name}: {marks} marks → PASS")
                    else:
                        failed_students.append({
                            "Name": name,
                            "Marks": marks,
                            "Status": "FAIL",
                            "File": filename
                        })
                        print(f"   {name}: {marks} marks → FAIL")

                except ValueError:
                    # Marks could not be converted to int
                    error_msg = f"Invalid marks for '{name}' in {filename} (line {line_number})"
                    errors_log.append(error_msg)
                    print(f"   {error_msg}")

                except IndexError:
                    # Row has missing columns
                    error_msg = f"Missing data in {filename} (line {line_number})"
                    errors_log.append(error_msg)
                    print(f"   {error_msg}")

    except FileNotFoundError:
        error_msg = f"File NOT found: {filename}"
        errors_log.append(error_msg)
        print(f"   {error_msg}")

    except ValueError as e:
        error_msg = f"Bad format in {filename}: {e}"
        errors_log.append(error_msg)
        print(f"   {error_msg}")

    except Exception as e:
        # Catch ANY other unexpected error
        error_msg = f"Unexpected error in {filename}: {e}"
        errors_log.append(error_msg)
        print(f"   {error_msg}")

print("\n" + "=" * 50)
print(" Processing complete! No crashes!")
print(f"   Passed: {len(passed_students)}")
print(f"   Failed: {len(failed_students)}")
print(f"   Errors: {len(errors_log)}")
```

```python
import csv

# STEP 1: Write clean results CSV

output_file = f"{folder}/clean_results.csv"

with open(output_file, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Marks", "Status", "Source File"])  # Header

    # Write passed students first
    for student in passed_students:
        writer.writerow([student["Name"], student["Marks"], student["Status"], student["File"]])

    # Then write failed students
    for student in failed_students:
        writer.writerow([student["Name"], student["Marks"], student["Status"], student["File"]])

print(f" Clean results saved to: {output_file}")


# STEP 2: Write error log

error_file = f"{folder}/error_log.txt"

with open(error_file, "w") as f:
    if errors_log:
        for error in errors_log:
            f.write(error + "\n")
    else:
        f.write("No errors found!\n")

print(f" Error log saved to: {error_file}")


# STEP 3: Display the clean results

print("\n" + "=" * 55)
print(" FINAL CLEAN RESULTS")
print("=" * 55)

with open(output_file, "r") as f:
    reader = csv.reader(f)
    header = next(reader)
    print(f"{'Name':<12} {'Marks':<8} {'Status':<8} {'File'}")
    print("-" * 55)
    for row in reader:
        print(f"{row[0]:<12} {row[1]:<8} {row[2]:<8} {row[3]}")


# STEP 4: Display the errors

print(f"\n{'=' * 55}")
print(" ERROR LOG")
print("=" * 55)

with open(error_file, "r") as f:
    for line in f:
        print(f"  • {line.strip()}")

# STEP 5: Show all files in the folder

print(f"\n All files in '{folder}/':")
for file in os.listdir(folder):
    size = os.path.getsize(f"{folder}/{file}")
    print(f"   {file} ({size} bytes)")
```

Task:
● Project Directory Auditor: The DevOps team needs a tool that scans a given project
folder, collects each file's name, size, type, and last modified date, then generates a
timestamped audit report. Files not modified in over 30 days must be flagged, and a
separate penalties report must list any calculated overage fines. Package the scanning
logic as a custom importable module that the audit script imports and uses.

```python


import os
from datetime import datetime, timedelta
import math



# STEP 1: Create the custom module file (scanner.py) using Python

# We use open() and write() to CREATE scanner.py automatically!
# This is our CUSTOM MODULE that we will import later.

scanner_code = '''
import os
from datetime import datetime, timedelta
import math


def scan_folder(folder_path):
    """Scans a folder and returns a list of file info dictionaries."""
    files_info = []
    items = os.listdir(folder_path)

    for item in items:
        full_path = os.path.join(folder_path, item)

        if os.path.isfile(full_path):
            file_stat = os.stat(full_path)
            size_bytes = file_stat.st_size
            last_modified = datetime.fromtimestamp(file_stat.st_mtime)
            _, extension = os.path.splitext(item)
            file_type = extension if extension else "No Ext"

            files_info.append({
                "name": item,
                "size_bytes": size_bytes,
                "file_type": file_type,
                "last_modified": last_modified
            })

    return files_info


def find_stale_files(files_info, days_limit=30):
    """Finds files not modified in the last 30 days."""
    stale_files = []
    today = datetime.now()

    for file in files_info:
        time_difference = today - file["last_modified"]

        if time_difference.days > days_limit:
            days_old = math.ceil(time_difference.days)
            stale_files.append({
                "name": file["name"],
                "last_modified": file["last_modified"],
                "days_since_modified": days_old
            })

    return stale_files


def calculate_penalties(stale_files, fine_per_day=0.50, days_limit=30):
    """Calculates fines for files over the day limit."""
    penalties = []

    for file in stale_files:
        overage_days = file["days_since_modified"] - days_limit
        fine = math.floor(overage_days * fine_per_day * 100) / 100

        penalties.append({
            "name": file["name"],
            "days_over_limit": overage_days,
            "fine": fine
        })

    return penalties


if __name__ == '__main__':
    print("Testing scanner module directly...")
    results = scan_folder(".")
    print(f"Found {len(results)} files.")
'''

# --- Write scanner.py file automatically ---
with open("scanner.py", "w") as f:
    f.write(scanner_code)

print("[OK] scanner.py module created automatically!")
print(f"     File exists: {os.path.exists('scanner.py')}")



# STEP 2: Import our custom module (scanner.py)

# Now we use "from ... import" to bring in our functions!

from scanner import scan_folder, find_stale_files, calculate_penalties

print("[OK] scanner module imported successfully!")
print("     Functions available: scan_folder, find_stale_files, calculate_penalties")



# STEP 3: Create sample project folder with test files

project_folder = "sample_project"

# os.path.exists() -- check if folder exists
# os.mkdir() -- create new folder
if not os.path.exists(project_folder):
    os.mkdir(project_folder)
    print(f"\n[+] Created folder: {project_folder}/")
else:
    print(f"\n[+] Folder already exists: {project_folder}/")

# Create 6 dummy files
sample_files = {
    "index.html": "<html><body>Hello World</body></html>",
    "style.css":  "body { color: black; background: white; }",
    "app.js":     "console.log('Hello from JS');",
    "readme.txt": "This is a sample project for auditing.",
    "data.csv":   "name,age,city\nAli,25,Lahore\nSara,30,Karachi",
    "notes":      "Some notes without extension",
}

for filename, content in sample_files.items():
    file_path = os.path.join(project_folder, filename)       # os.path.join!
    with open(file_path, "w") as f:
        f.write(content)

print(f"[OK] Created {len(sample_files)} sample files")

# --- Make 2 files look OLD using timedelta (date arithmetic) ---
old_files_days = {"readme.txt": 45, "notes": 60}

for old_file, days in old_files_days.items():
    old_path = os.path.join(project_folder, old_file)
    old_time = datetime.now() - timedelta(days=days)         # timedelta!
    old_timestamp = old_time.timestamp()
    os.utime(old_path, (old_timestamp, old_timestamp))
    print(f"    Made '{old_file}' look {days} days old")



# STEP 4: Scan the folder (using IMPORTED function!)

print("\nScanning folder...")
files_info = scan_folder(project_folder)        # from scanner.py!
print(f"    Found {len(files_info)} files")



# STEP 5: Find stale files (using IMPORTED function!)

stale_files = find_stale_files(files_info, days_limit=30)   # from scanner.py!
print(f"    Found {len(stale_files)} stale files (30+ days old)")



# STEP 6: Print Audit Report

now = datetime.now()

print("\n" + "=" * 55)
print("        PROJECT DIRECTORY AUDIT REPORT")
print("=" * 55)
print(f"Folder Scanned : {project_folder}")
print(f"Report Date    : {now.strftime('%A, %B %d, %Y')}")       # strftime!
print(f"Report Time    : {now.strftime('%I:%M:%S %p')}")          # strftime!
print(f"Total Files    : {len(files_info)}")
print(f"Stale Files    : {len(stale_files)} (not modified in 30+ days)")
print("=" * 55)

# Print file details table
print(f"\n{'File Name':<18} {'Size(B)':<9} {'Type':<10} {'Last Modified':<18}")
print("-" * 55)

for file in files_info:
    mod_date = file["last_modified"].strftime("%d-%m-%Y %H:%M")  # strftime!
    print(f"{file['name']:<18} {file['size_bytes']:<9} {file['file_type']:<10} {mod_date:<18}")

# Print stale files section
print("\n" + "=" * 55)
print("STALE FILES (Not modified in 30+ days):")
print("-" * 55)

if stale_files:
    for sf in stale_files:
        mod_date = sf["last_modified"].strftime("%d-%m-%Y")      # strftime!
        print(f"  [!] {sf['name']} -- Modified: {mod_date} ({sf['days_since_modified']} days ago)")
else:
    print("  All files are up to date!")

print("=" * 55)



# STEP 7: Calculate penalties (using IMPORTED function!)

penalties = calculate_penalties(stale_files, fine_per_day=0.50, days_limit=30)  # from scanner.py!



# STEP 8: Print Penalties Report

print("\n" + "=" * 55)
print("        PENALTIES REPORT -- OVERAGE FINES")
print("=" * 55)
print(f"Generated : {now.strftime('%d-%m-%Y at %I:%M %p')}")    # strftime!
print(f"Fine Rate : $0.50 per day over 30-day limit")
print("-" * 55)

total_fine = 0

if penalties:
    print(f"{'File Name':<18} {'Days Over':<12} {'Fine ($)':<10}")
    print("-" * 55)

    for p in penalties:
        print(f"{p['name']:<18} {p['days_over_limit']:<12} ${p['fine']:<10.2f}")
        total_fine = total_fine + p["fine"]

    print("-" * 55)
    print(f"{'TOTAL FINE:':<30} ${total_fine:.2f}")
else:
    print("No penalties! All files are within the 30-day limit.")

print("=" * 55)


# STEP 9: Verify created files using os module

print("\n" + "=" * 55)
print("VERIFICATION -- Files Created:")
print("-" * 55)
print(f"  scanner.py exists  : {os.path.exists('scanner.py')}")           # os.path.exists!
print(f"  sample_project/    : {os.path.exists('sample_project')}")
print(f"  Files in folder    : {os.listdir('sample_project')}")           # os.listdir!
print("=" * 55)

print("\n[DONE] Audit complete! All results shown above.")
```

HR Employee Ranker: HR wants employees ranked by performance score, then by
experience as a tiebreaker, then by name alphabetically. The system must also support
quickly searching for a specific employee by name from the sorted list.

```python
# 1. HR ka Data (Kachra/Unsorted)
employees = [
    {"name": "Zain", "score": 85, "experience": 3},
    {"name": "Ali", "score": 90, "experience": 5},
    {"name": "Sara", "score": 85, "experience": 4},
    {"name": "Ahmed", "score": 90, "experience": 5},
    {"name": "Bilal", "score": 70, "experience": 2}
]


# STEP 2: Tiebreaker Sorting (Multiple Keys)

-# Hukm (Logic):
# 1. Score zyada ho (Descending)
# 2. Agar score same ho, toh Experience zyada ho (Descending)
# 3. Agar dono same hon, toh Naam A-Z ho (Ascending)
# Trick: Numbers ke peechay '-' (minus) lagane se wo automatically ulta (High to Low) sort ho jate hain!

sorted_employees = sorted(
    employees,
    key=lambda x: (-x["score"], -x["experience"], x["name"])
)


# STEP 3: O(1) Fast Search Index (Dict Comprehension + Enumerate)

# Chuke list 'Score' ke hisab se sort hui hai (Naam ke hisab se nahi),
# isliye hum Binary Search nahi laga sakte.
# Iska sab se tez O(1) hal yeh hai ke hum sorted data ko ek Dictionary mein badal lein.

search_index = {
    emp["name"]: {"rank": index, "details": emp}
    for index, emp in enumerate(sorted_employees, start=1)
}


# STEP 4: Results Dekhein

print(" --- HR RANKING BOARD --- ")
for index, emp in enumerate(sorted_employees, start=1):
    print(f"Rank {index}: {emp['name']:<6} | Score: {emp['score']} | Exp: {emp['experience']} yrs")

print("\n --- QUICK EMPLOYEE SEARCH --- ")
# HR kisi ka naam likh kar direct uski rank nikal sakti hai
search_name = "Bilal"

if search_name in search_index:
    found = search_index[search_name]
    print(f" Found {search_name} Rank {found['rank']}  (Score: {found['details']['score']})")
else:
    print(f"Not Found! the record  of {search_name} does not exist.")
```

Task:
● Order Discount Pipeline: An online store needs a data pipeline that takes a raw list of
customer orders, applies category-based discount rules, filters out orders below a
minimum threshold, and returns a processed summary with total revenue — using no
traditional `for` loops anywhere.

```python
orders = [
    {"id": 1, "category": "electronics", "price": 1000},
    {"id": 2, "category": "clothing", "price": 300},
    {"id": 3, "category": "electronics", "price": 400},
    {"id": 4, "category": "grocery", "price": 200}
]
```

```python
discount_rules = {
    "electronics": 0.10,
    "clothing": 0.20,
    "grocery": 0.0
}  #discount rules
```

```python
discounted_orders = list(
    map(
        lambda order: {
            **order,
            "final_price": order["price"] - (order["price"] * discount_rules[order["category"]])
        },
        orders
    )
)  #applying discount using map function
```

```python
filtered_orders = list(
    filter(
        lambda order: order["final_price"] >= 300,
        discounted_orders
    )
) #minimum threshhold applying
```

```python
from functools import reduce

total_revenue = reduce(
    lambda total, order: total + order["final_price"],
    filtered_orders,
    0
)#total revnue calculated
```

```python
print("Processed Orders:")
print(filtered_orders)

print("\nTotal Revenue:")
print(total_revenue)
```

```python
import time

# 1. File Reader Generator (read one line at a time )
def read_logs_stream(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip() # Sirf current line memory mein aayegi

# 2. Filter Generator (only error and critical )
def filter_errors(log_stream):
    for line in log_stream:
        if "ERROR" in line or "CRITICAL" in line:
            yield line

# 3. Error Type Extractor (Name of the error)
def extract_error_type(error_stream):
    for line in error_stream:
        # Example: "ERROR: TimeoutError - Database..." ko split karke "TimeoutError" nikalna
        parts = line.split(":")
        if len(parts) > 1:
            error_type = parts[1].split("-")[0].strip()
            yield error_type

# --- MAIN EXECUTION (Live Running Count) ---

print("\n--- Live Error Stream Processing Shuru ---")

# Apna Generator pipeline set karein
logs = read_logs_stream("server_logs.txt")
errors_only = filter_errors(logs)
error_types_stream = extract_error_type(errors_only)

# Live count store karne ke liye ek dictionary
error_counter = {}

# Stream ko run karein aur live count dikhayein
for error in error_types_stream:
    # Agar error pehle se count mein hai toh +1 karein, warna 1 set karein
    if error in error_counter:
        error_counter[error] += 1
    else:
        error_counter[error] = 1

    # Live update print karein
    print(f" New Error Comes: {error} | Total Count: {error_counter[error]}")

    # Thora delay daal rahe hain taake live streaming ka feel aaye
    time.sleep(0.5)

print("\n--- Log Processing Completed  ---")
```

● API Middleware Simulator: A backend team needs reusable function wrappers that
handle execution timing, input/output logging, and role-based access control — without
touching the original functions. All wrappers must be stackable and all activity must go
through the `logging` module, not print.

```python
import logging

# Logger create
logger = logging.getLogger("api_middleware")
logger.setLevel(logging.DEBUG)

# Console handler
console_handler = logging.StreamHandler()
console_handler.setLevel(logging.INFO)

# File handler
file_handler = logging.FileHandler("api.log")
file_handler.setLevel(logging.DEBUG)

# Formatter
formatter = logging.Formatter(
    "%(asctime)s | %(levelname)s | %(message)s"
)

console_handler.setFormatter(formatter)
file_handler.setFormatter(formatter)

logger.addHandler(console_handler)
logger.addHandler(file_handler)
```

```python
# Decorator – Execution Timer
import time
from functools import wraps

def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        logger.info(f"{func.__name__} executed in {end - start:.4f} seconds")
        return result
    return wrapper
```

```python
# Decorator – Input / Output Logging
def log_io(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        logger.debug(f"{func.__name__} INPUT args={args}, kwargs={kwargs}")
        result = func(*args, **kwargs)
        logger.debug(f"{func.__name__} OUTPUT result={result}")
        return result
    return wrapper
```

```python
# Decorator – Role-Based Access Control
def require_role(required_role):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            role = kwargs.get("role")

            if role != required_role:
                logger.warning(
                    f"ACCESS DENIED to {func.__name__} | role={role}"
                )
                return {"error": "Access Denied"}

            logger.info(
                f"ACCESS GRANTED to {func.__name__} | role={role}"
            )
            return func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
# Original API Function (UNCHANGED )
@timer
@log_io
@require_role("admin")
def create_user(username, email, role=None):
    time.sleep(1)  # simulate processing
    return {
        "status": "success",
        "username": username,
        "email": email
    }
```

Task:
● Banking System: Build a `BankAccount` class that stores an owner name and balance
securely. The balance must be inaccessible from outside the class. Deposits and
withdrawals go through validated methods only. Print any account and get a clean,
readable summary.

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner              # public attribute
        self.__balance = balance        # private attribute

    # Deposit method
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposit successful: {amount}")
        else:
            print("Deposit amount must be positive")

    # Withdraw method
    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be positive")
        elif amount > self.__balance:
            print("Insufficient balance")
        else:
            self.__balance -= amount
            print(f"Withdrawal successful: {amount}")

    # Read-only balance access
    @property
    def balance(self):
        return self.__balance

    # User-friendly output
    def __str__(self):
        return f"BankAccount Owner: {self.owner}, Balance: {self.__balance}"

    # Developer-friendly output
    def __repr__(self):
        return f"BankAccount(owner='{self.owner}', balance={self.__balance})"
```

```python
          #  Testing
# Object create karna
account1 = BankAccount("Ali", 1000)

# Deposit
account1.deposit(500)

# Withdraw
account1.withdraw(300)

# Invalid cases
account1.withdraw(5000)
account1.deposit(-100)

# Balance check (secure way)
print("Current Balance:", account1.balance)

# Print account summary
print(account1)
```

Task:
● Library Management System: A library needs to manage books, members, and
borrowing records. Books are borrowed and returned. Members have limits. Overdue
books trigger fines. Different membership tiers have different rules and privileges. Use
single and multilevel inheritance for tiers, method overriding for fine logic, and dunder
methods for display.

```python
# -------------------- Book Class --------------------
class Book:
    def __init__(self, title):
        self.title = title
        self.is_borrowed = False

    def __str__(self):
        status = "Borrowed" if self.is_borrowed else "Available"
        return f"Book: {self.title} | Status: {status}"


# -------------------- Member Base Class --------------------
class Member:
    def __init__(self, name):
        self.name = name
        self.borrow_limit = 2
        self.borrowed_books = []

    def borrow_book(self, book):
        if len(self.borrowed_books) >= self.borrow_limit:
            print(f"{self.name}  borrow limit exceed!")
            return
        if book.is_borrowed:
            print("Book already borrowed.")
            return

        book.is_borrowed = True
        self.borrowed_books.append(book)
        print(f"{self.name} borrow '{book.title}'  book.")

    def return_book(self, book, days_late=0):
        if book in self.borrowed_books:
            book.is_borrowed = False
            self.borrowed_books.remove(book)
            fine = self.calculate_fine(days_late)
            print(f"{self.name} return '{book.title}' book. Fine: Rs {fine}")
        else:
            print("Ye book is member ke paas nahi hai.")

    def calculate_fine(self, days_late):
        return days_late * 10  # base fine logic

    def __len__(self):
        return len(self.borrowed_books)

    def __str__(self):
        return f"Member: {self.name} | Borrowed Books: {len(self)}"


# -------------------- Silver Member (Single Inheritance) --------------------
class SilverMember(Member):
    def __init__(self, name):
        super().__init__(name)
        self.borrow_limit = 3

    def calculate_fine(self, days_late):
        return days_late * 7  # overridden fine logic


# -------------------- Gold Member (Multilevel Inheritance) --------------------
class GoldMember(SilverMember):
    def __init__(self, name):
        super().__init__(name)
        self.borrow_limit = 5

    def calculate_fine(self, days_late):
        return days_late * 3  # lowest fine


# -------------------- Testing the System --------------------

# Books
book1 = Book("Python Basics")
book2 = Book("OOP in Depth")
book3 = Book("Data Science")
book4 = Book("AI for Beginners")

# Members
normal = Member("Ali")
silver = SilverMember("Sara")
gold = GoldMember("Ahmed")

print("\n--- Borrowing Books ---")
normal.borrow_book(book1)
normal.borrow_book(book2)
normal.borrow_book(book3)   # limit exceed

silver.borrow_book(book3)
gold.borrow_book(book4)

print("\n--- Returning Books with Delay ---")
normal.return_book(book1, days_late=2)
silver.return_book(book3, days_late=2)
gold.return_book(book4, days_late=0)

print("\n--- Display using Dunder Methods ---")
print(book1)
print(normal)
print(silver)
print(gold)
```
